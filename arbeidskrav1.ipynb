## Arbeidskrav 1

Anta at du skal kjøpe bil. Det står mellom elbil og bensinbil, og du ønsker å sammenlikne de årlige kostnadene ved elbil sammenliknet med bensinbil.

Lag et Python-program som beregner og presenterer (i konsollen) de årlige totalkostnadene for elbil og for bensinbil samt årlig kostnadsdifferanse basert på informasjonen gitt nedenfor. Du kan her for enkelhets skyld se bort fra kostnader som renter på billån og verditap (du har da egentlig antatt at slike kostnader er like for elbil og bensinbil).

Nedenfor er informasjon som programmet skal baseres på (som selvsagt kan diskuteres, men ikke ifm. denne oppgaven :-)

Du kan selv velge antall kjørte km/år ut fra din typiske bilbruk. Ev. (hvis du ikke har bil) kan du anta 10.000 km. Forsikring: Elbil: 5000 kr/år. Bensinbil: 7500 kr/år. Trafikkforsikringsavgift: 8,38 kr/dag for både elbil og bensinbil. Drivstoffbruk: Elbil: 0,2 kWh/km. Strømpris (antar kun hjemmelading): 2.00 kr/kWh. Bensinbil: 1,0 kr/km. Bomavgift: Elbil: 0,1 kr/km. Bensinbil: 0,3 kr/km.

```python
'''
Arbeidskrav 1
Program for beregning av årlige kjøre kostnader
for henholdsvis elbil og bensinbil
'''
kjørelengde = 10_000    # Kjørelengde km/år 
trafikk_avgift = 8.38   # Trafikkforsikringsavgift for elbil og bensinbil : 8,38 kr/dag 
strømpris = 2.00        # Strømpris: 2.00 kr/kWh
dager_i_året = 365      # Dager i et år 
forsikring_eb = 5000    # Forsikring elbil: 5000 kr/år
forsikring_bb = 7500    # Forsikring bensinbil: 7500 kr/år.
drivstoff_eb = 0.2      # drivstoffbruk elbil: 0,2 kWh/km
drivstoff_bb = 1.0      # drivstoffbruk bensinbil: 1,0 kr/km
bomavgift_eb = 0.1      # Bomavgift elbil: 0,1 kr/km
bomavgift_bb = 0.3      # Bomavgift bensinbil: 0,3 kr/km.

kostnad_eb = (trafikk_avgift * dager_i_året + 
              forsikring_eb + 
              drivstoff_eb * strømpris * kjørelengde +
              bomavgift_eb * kjørelengde
)

kostnad_bb = (trafikk_avgift * dager_i_året + 
              forsikring_bb +
              drivstoff_bb * kjørelengde +
              bomavgift_bb * kjørelengde
)

differanse = abs(kostnad_eb - kostnad_bb)
print(f'Årlige kostnader elbil....: {kostnad_eb:.2f} kr')
print(f'Årlige kostander bensinbil: {kostnad_bb:.2f} kr')
print(f'Differanse................: {differanse:.2f} kr')
```
